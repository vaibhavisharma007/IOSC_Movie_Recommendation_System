```python
# ------imp-features-------
#genres
#id(we have to fetch posters through id)
#keywords(for tags)
# title
# overview
# cast
# crew
new_df=movies[["id","title","overview","genres","keywords","cast","crew"]]
new_df.head()
```

```python
import ast
def dict_to_list(listofdict):
  list_=[]
  for dict_ in ast.literal_eval(listofdict):
    list_.append(dict_["name"])
  return list_
```

```python
new_df["genres"]=new_df["genres"].apply(dict_to_list)
```

```python
new_df.head()
```

```python
new_df.iloc[0]["keywords"]
```

```python
new_df["keywords"]=new_df["keywords"].apply(dict_to_list)
```

```python
new_df.iloc[0]["cast"]
```

```python
import ast
def convert_cast(listofdict):
  list_=[]
  count=0
  for dict_ in ast.literal_eval(listofdict):
    count+=1
    list_.append(dict_["name"])
    if(count==3):
      return list_
  return list_
```

```python
new_df["cast"]=new_df["cast"].apply(convert_cast)
```

```python
new_df.head()
```

```python
new_df.iloc[0]["crew"]
```

```python
import ast

def fetch_director(data):
    try:
        crew_data = ast.literal_eval(data)  # Convert string to list of dictionaries
        directors = [crew['name'] for crew in crew_data if crew.get('job') == 'Director']
        return directors
    except (SyntaxError, ValueError):
        return None  # Handle invalid data gracefully

# Apply the function to the DataFrame
new_df["crew"] = new_df["crew"].apply(fetch_director)
```

```python
new_df.head()
```

```python
new_df.iloc[0]["overview"]
```

```python
new_df["overview"] = new_df["overview"].apply(lambda x: x.split())
```

```python
new_df["crew"].dropna(inplace=True)
```

```python
new_df["genres"]=new_df["genres"].apply(lambda x:[i.replace(" ","") for i in x])
new_df["overview"]=new_df["overview"].apply(lambda x:[i.replace(" ","") for i in x])
new_df["keywords"]=new_df["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
new_df["cast"]=new_df["cast"].apply(lambda x:[i.replace(" ","") for i in x])
new_df["crew"] = new_df["crew"].apply(lambda x: [i.replace(" ", "") for i in x] if x is not None else [])
```

```python
new_df["tag"]=new_df["overview"]+new_df["genres"]+new_df["keywords"]+new_df["crew"]+new_df["cast"]
```

```python
new_df.head()
```

```python
df=new_df.drop(columns=["overview","genres","keywords","cast","crew"])
```

```python
df.head()
```

```python
df["tag"]=df["tag"].apply(lambda x:" ".join(x))
```

```python
df["tag"]=df["tag"].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vector=cv.fit_transform(df["tag"]).toarray()
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
# ps convert some same typ of word like love loving lovable etc into root word as love
```

```python
df["tag"]=df["tag"].apply(stem)
```

```python
vector=cv.fit_transform(df["tag"]).toarray()
```

```python
series=cv.get_feature_names_out()
# data=pd.DataFrame(data=series)
# data.value_counts()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
cosine_similarity(vector).shape
```

```python
similarity[0]
```

```python
# diagnol of similarity will be one
```

```python
def recommend(movie):
  i=df[df["title"]==movie].index[0]
  dist=similarity[i]
  movie_list=sorted(list(enumerate(dist)),reverse=True,key=lambda x:x[1])[1:6]


  l=[]
  for i in movie_list:

    l.append(df.loc[i[0]]["title"])
  return l
```

```python
recommend("Captain America: Civil War")
```

```python
 for i in recommend("Avatar"):
  print(df.loc[i]["title"])
```
